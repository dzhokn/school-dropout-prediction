## Linear Regression
Linear regression is a statistical method used to model the relationship between an input variable $X$ and output variable $Y$.


In other words linear regression produces a **linear function** :
$$f(x)=b + wx$$
that maps the data points in the best possible way. In result that linear function can be used for prediction on new datasets.

For example, let's say we have a dataset where every row represents the *salary* of a data scientist and her corresponding *years of experience*.

```python
# Let's name the years of experience 'X' (an input variable)
X = [1.2, 1.3, 1.5, 1.8, 2, 2.1, 2.2, 2.5, 2.8, 2.9, 3.1, 3.3, 3.5, 3.8, 4, 4.1, 4.5, 4.9, 5, 5.5, 6, 6.5, 7, 7.5, 8, 8.5, 9, 10, 11, 12, 13, 14, 15]

# And let's name the corresponding salaries 'Y' (the target/output variable)
Y = [2900, 3300, 3100, 4200, 3500, 3800, 3300, 3500, 3750, 4000, 3900, 5300, 4420, 5000, 4900, 5200, 3900, 4800, 5700, 6500, 6930, 7500, 7360, 6970, 6800, 7500, 8000, 9500, 11000, 9500, 12300, 13700, 12500]
```

Let's plot the data to see how it is placed in a 2-dimensional space.

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Now, can we come up with a linear, polynomial function $$f(x) = b + w x$$ that represents the above data as close as possible.

But what does *"as close as possible"* mean? How can we measure it?

#### Cost function
The cost function measures the difference between predicted values and actual outcomes. It is also called *error function* or *loss function*.

For example, let's calculate the cost (the deviation) for the function $$f(x) = 2000 + 800 x$$

```python
import numpy as np

X = np.vectorize(float)(X)
plt.plot(X, 2000 + 800 * X, color='g')
plt.scatter(X, Y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

### Mean Squared Error
It looks pretty close visually. Let's measure it precisely now using the most common formula for cost function - **Mean Squared Error (MSE)**.

$$ J = \frac{1}{m} \sum_{i=1}^{m} (f_{w,b}(x^{(i)}) - y^{(i)})^2 $$

Where $i$ is simply the iterator for all rows in our dataset ($m$ = $33$ rows in total).

$w$, $b$ are the parameters of our linear function $f(x)$ - the one that is expected to represent the data we have.

$y^{(i)}$ are the actual expected results, defined in the `Y` dataset.

This cost function measures the mean squared error (MSE). The higher power is used to make the cost function more sensitive to errors and more importantly to make it **differentiable**.

#### Why do we prefer MSE over MAE?
Mean Squared Error (MSE) is often preferred over Mean Absolute Error (MAE) because it's differentiable everywhere, allowing for easier optimization using gradient-based algorithms.

```python
def compute_cost(X: np.ndarray, Y: np.ndarray, w: float, b: float) -> float:
    m = len(X)
    cost = 0
    for i in range(m):
        cost += (w * X[i] + b - Y[i]) ** 2
    return cost / m

print(compute_cost(X, Y, 2000, 800))
```
